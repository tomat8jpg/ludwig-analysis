```python
import pandas as pd
```

```python
df = pd.read_csv('Desktop/что-то нужное или нет/прога/vgsales.csv')
```

```python
df.head()
```

```python
from ludwig.api import LudwigModel
```

```python
model_definition = {
    'input_features':[
        {'name': 'Genre', 'type': 'category'},
        {'name': 'Publisher', 'type': 'category'},
        {'name': 'NA_Sales', 'type': 'numerical'}
    ],
    'output_features':[
        {'name': 'Global_Sales', 'type': 'numerical'}
    ]
}
```

```python
model = LudwigModel (model_definition)
train_stats = model.train(df)
```

```python
train_stats
```

```python
from ludwig.visualize import *
```

```python
learning_curves(train_stats, 'Global_Sales')
```
